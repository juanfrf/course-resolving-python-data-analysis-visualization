## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=groceries, y=price, c='black', marker='*')
```

### Bar plot

```python
plt.bar(x=groceries, height=price, width=0.7, align='center')
```

### Many others

https://matplotlib.org/stable/plot_types/index

## Customize the plot

### Change the size

```python
plt.scatter(x=groceries,y=price,s=100)
```

### Change the color

```python
plt.scatter(x=groceries,y=price,s=500,c='black')
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x=groceries,y=price,marker='s')
```

### Change the border

```python
plt.scatter(x=groceries,y=price, c='white',edgecolors='black')
```
