# Python Foundations for Data Visualization

## Create objects with data

### Football Teams

```python
football_teams = ['Real Madrid', 'Barcelona', 'Bayern Munich']
```

### Champions Leagues

```python
champions_leagues = [14, 5, 6]
```

## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=football_teams,y=champions_leagues)
```

### Bar plot

```python
plt.bar(x=football_teams,height=champions_leagues,width=0.7)
```

### Many others

https://matplotlib.org/stable/plot_types/index

## Customize the plot

### Change the size

```python
plt.scatter(x=football_teams,y=champions_leagues, s=100)
```

### Change the color

```python
plt.scatter(x=football_teams,y=champions_leagues,c='red')
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x=football_teams,y=champions_leagues,c='yellow',marker='*')
```

### Change the border

```python
plt.scatter(x=football_teams,y=champions_leagues,c='white',edgecolors='black')
```
